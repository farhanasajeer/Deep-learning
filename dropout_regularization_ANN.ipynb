Dropout regularization in Deep nueral network

This is dataset that describes sonar chirp returns bouncing different services.The 60 input variables are the strength of the return at different angles. It is a binary classification problem that requires a model from metal cylinders.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/drive/MyDrive/sonar_dataset.csv',header=None)
df.sample(5)
```

```python
df.shape
```

```python
# check for nan values
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
# label is not skewed
```

```python
X=df.drop(60,axis=1)
y=df[60]
y.head()
```

```python
y=pd.get_dummies(y, drop_first=True)
y.sample(5)


# r=1 m=0
```

```python
y.value_counts()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train.head()
```

USING DEEP LEARNING MODEL

MODEL WITHOUT DROPOUT LAYER

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
                          keras.layers.Dense(60, input_dim=60, activation='relu'),
                          keras.layers.Dense(30, activation='relu'),
                          keras.layers.Dense(15, activation='relu'),
                          keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
# Training accuracy >> test accuracy
y_pred=model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

```python
# model dropout layer

model = keras.Sequential([
                          keras.layers.Dense(60, input_dim=60, activation='relu'),
                          keras.layers.Dropout(0.5),
                          keras.layers.Dense(30, activation='relu'),
                          keras.layers.Dropout(0.5),
                          keras.layers.Dense(15, activation='relu'),
                          keras.layers.Dropout(0.5),
                          keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
# Training accuracy is still good test accuracy improved
y_pred=model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```
